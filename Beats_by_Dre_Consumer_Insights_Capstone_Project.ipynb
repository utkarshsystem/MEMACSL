## 6. Sentiment Analysis

### 6.1/6.2 Methodology & Results
Textblob processes textual data and helps perform natural language processing (NLP) tasks. Sentiment analysis in TextBlob is based on the polarity of the text, which ranges from -1 (negative) to 1 (positive).
The sentiment polarity score is calculated for each text:
- Positive Polarity (> 0) indicates positive sentiment.
- Neutral Polarity (= 0) indicates neutral sentiment.
- Negative Polarity (< 0) indicates negative sentiment.

```python
# Part 4: Sentiment Analysis using TextBlob:
from textblob import TextBlob

# Due to 'content' column error
def get_sentiment(review):
    # Handle non-string values
    if not isinstance(review, str):
        return 'N/A'

    analysis = TextBlob(review)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

df['Sentiment'] = df['content'].apply(get_sentiment)
df.head()
```

The mainly positive reviews suggests the customers were generally satisfied with the Beats by Dre speakers, however there are reviews that show places for improvement. ### 8. SWOT Analysis
Strengths, Weaknesses, Opportunities, Threats for Beats by Dre based on the comparative data.

Based on all the data from comparing Beats and other products speakers on Amazon, I have created a SWOT analysis for Beats:
- Strengths: The design and quality of sound is well known and lives up to standard in the audio industry. The design and branding are perceived as modern and popular amongst other brands. There is a strong bass its integration with Apple offers discounts once in a while.
- Weaknesses: According to the data of reviews, some weaknesses include pricing and customer service. People consider it overpriced compared to JBL and Sony. Some negative feedback included customer support not being responsive. Alongside these issues was battery life which people found not long enough compared to speakers such as Anker and JBL.
- Opportunities: Some innovations on the technology side include improving battery life and further innovations when it comes to the sound quality of the speakers. Some improvements on the marketing side could be more pressence on social media but specifically to  younger audiences as that is a big part of the customer audience.
- Threats: Competitors like Sony, JBL, and Bose are consistently said to deliver better sound quality, battery life, and durability than Beats. Those and Anker are currently the most competitive brands for Beats. It is important to address the negative comments because negative feedback related to customer support and product reliability could lead to reduced consumer trust which effects the brand loyalty.
